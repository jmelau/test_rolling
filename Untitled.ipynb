```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

# Load the data.
test = pd.read_excel("test_ROLLING.xlsx")
test.head()
```

Time_water is athletes swimming. Time_water is time from start. Datacollection from aprox 15 min before start
Time_land is athletes out of the water and cycling. 

```python
plt.figure(figsize=(14,9))
# Overal size of the figure 

plt.style.use('seaborn')

plt.title('Non rolling figure', fontdict={'fontweight':'bold', 'fontsize': 20})
# Title of the plot.

# All lines plottet individually. 
# Separate plots for in the water and out of the water(land)
# tp = temperature
plt.plot(test.time_water_02, test.tp_water_02, 'b')
plt.plot(test.time_land_02, test.tp_water_02, 'g')
plt.plot(test.time_water_04, test.tp_water_04, 'b')
plt.plot(test.time_land_04, test.tp_land_04, 'g')


plt.xlabel('Race time in minutes')
plt.ylabel('Core temperature')
# Title of x- and y-labels

 # Limits for plot. Should be the same for all plots
plt.xlim()
plt.ylim(36,39)

plt.tight_layout()
# Gir en litt annen layout. 


plt.show()
```

Plot above is with raw data. 

Plot below is moving average
Need this plot to be setup as plot above.. 

```python
test.rolling(3, on='time_water_02').mean().plot(x='time_water_02', ylim=(36,39))
```
